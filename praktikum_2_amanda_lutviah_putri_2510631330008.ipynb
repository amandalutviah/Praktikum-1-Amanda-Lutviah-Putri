```python
import matplotlib.pyplot as plt

g = 9.8 #percepatan gravitasi (m**2)
h = 100 #ketinggian awal (m)
Vo = 0 #kecepatan awal (m/s)
dt = 1 #interval waktu (s)
t = list(range(11))
s = [0.5*g*(ti**2) for ti in t]
v = [g* ti for ti in t]
a = [g] * len(t)



plt.plot(t,s, marker='o', label='Posisi')
plt.plot(t,v, marker='o', label='Kecepatan')
plt.plot(t,a, marker='o', label='Percepatan')
plt.title("Grafik posisi vs Waktu: Gerak jatuh bebas")
plt.xlabel("Waktu (detik)")
plt.ylabel("Posisi (m)")
plt.grid(True)
plt.legend()
plt.show()

#Nlai maksimum
print("kasus 1:")
print(f"posisi maksimum: {max(s):.2f}m")
print(f"kecepatan maksimum: {max(v):.2f}m/s")
print(f"percepatan maksimum: {max(a):.2f}m/sÂ²")
```
