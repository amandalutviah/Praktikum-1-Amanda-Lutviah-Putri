```
import numpy as np
import matplotlib.pyplot as plt
import math

def shm_position(amplitude, omega, time):
  return amplitude * math.cos(omega*time)

print ("Simulasi Gerak Harmonik Sederhana (SHM)")
amplitude = float(input("masukan amplitude (A) dalam meter:"))
omega = float(input("masukan frekuensi sudut (omega) dalam rad/s:"))
total_time = float(input("masukan Waktu total simulasi dalam detik :"))
time_step = 0.1 # Langkah Waktu (dt)

times = []
positions = []

current_time = 0.0
while current_time <= total_time :
  pos = shm_position(amplitude, omega,current_time)

  #if condition : Jika posisi melebihi amplitude meskipun tidak mungkin dalam SHM ideal, ini untuk demo)
  if abs(pos) > amplitude * 1.1 :
    #kondisi batas (misalnya, jika ada error numerik)
    print (f"Simulasi dihentikan pada t={current_time : .2f}s karena posisi melebihi batas.")
    break
  times.append(current_time)
  positions.append(pos)
  current_time += time_step

plt.figure(figsize=(10,6))
plt.plot(times, positions, label='posisi vs waktu',color='blue', linewidth = 2)
plt.title('Simulasi Gerak Harmonik Sederhana')
plt.xlabel('waktu(s)')
plt.ylabel('posisi (m)')
plt.grid(True)
plt.legend()
plt.show()

print(f"Simulasi selesai. Total titik data : {len(times)}")
```
